**Import library's**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Load images and show**

```python
flat_chess = cv2.imread('chessboard.png')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray_flat_chess, cmap='gray')
```

```python
gray_flat_chess
```

```python
gray = np.float32(gray_flat_chess)
```

Harris corner detection looks at two different gradients that are incoming. How far off, does the eigen vector has to be, to be a corner versus not a corner? That is the 'k' value. How far the eigen values of gradients needs to be? 'ksize' is a parameter needed for the sobel gradient calculation.

```python
dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.03) 
```

```python
dst = cv2.dilate(dst, None) 
```

```python
# This process is for visualization
flat_chess[dst > 0.01 * dst.max()] = [255, 0, 0] # RGB
```

```python
plt.imshow(flat_chess)
```
