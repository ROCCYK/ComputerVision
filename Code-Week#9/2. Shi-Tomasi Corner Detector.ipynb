**Import libary's**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Load the images of chess board**

```python
flat_chess = cv2.imread('chessboard.png')
#flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2GRAY)
```

**Called the good features the track function to find 5 corners**

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess, maxCorners=50, qualityLevel=0.01, minDistance=10)
```

```python
#corners
```

```python
corners = np.uint0(corners)
```

```python
#corners
```

```python
for i in corners:
    x, y = i.ravel()
    cv2.circle(flat_chess, (x, y), 3, (255, 0, 0), -1)
```

```python
plt.imshow(flat_chess)
```

**Called the good features the track function to find all corners**

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess, 64, 0.01, 10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x, y = i.ravel()
    cv2.circle(flat_chess, (x, y), 3, (255, 0, 0), -1)
```
